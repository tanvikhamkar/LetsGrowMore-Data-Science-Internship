# Importing Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

# Original Image

```python
image = cv2.imread("C:\\Users\\Lenovo\\Desktop\\mountain-flowers-nature.jpg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

# Grayscale Image

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

# Negative Image

```python
negative_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(negative_image))
```

# Blurry Image

```python
blurry_image = cv2.GaussianBlur(negative_image,(21,21),0)
display(Image.fromarray(blurry_image))
```

# Inverted Blurry Image

```python
inverted_blur = cv2.bitwise_not(blurry_image)
display(Image.fromarray(inverted_blur))
```

# Pencil Sketch

```python
sketch = cv2.divide(gray_image,inverted_blur,scale=256.0)
display(Image.fromarray(sketch))
```
